# Vector Operations

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

Ax, Ay = 0.5, -1
Bx, By = 1, 1.5

def style_ax(ax, title):
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    ax.set_xlim(-4, 4)
    ax.set_ylim(-3, 4)
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.set_title(title, pad=20, fontweight='bold', fontsize=14)


style_ax(ax1, "Addition: A + B")

ax1.quiver(0, 0, Ax, Ay, angles='xy', scale_units='xy', scale=1, color='#4A86E8', width=0.008)
ax1.quiver(0, 0, Bx, By, angles='xy', scale_units='xy', scale=1, color='#34A853', width=0.008)

Rx, Ry = Ax + Bx, Ay + By
ax1.quiver(0, 0, Rx, Ry, angles='xy', scale_units='xy', scale=1, color='#F6B26B', width=0.008)

ax1.plot([Ax, Rx], [Ay, Ry], linestyle='--', color='#34A853', alpha=0.6, linewidth=2)
ax1.plot([Bx, Rx], [By, Ry], linestyle='--', color='#4A86E8', alpha=0.6, linewidth=2)

ax1.text(Ax/2 - 0.4, Ay/2, 'A', color='#4A86E8', fontweight='bold', fontsize=12)
ax1.text(Bx/2, By/2 - 0.3, 'B', color='#34A853', fontweight='bold', fontsize=12)
ax1.text(Rx/2 + 0.1, Ry/2, 'A+B', color='#F6B26B', fontweight='bold', fontsize=12)



style_ax(ax2, "Substraction: A - B")

ax2.quiver(0, 0, Ax, Ay, angles='xy', scale_units='xy', scale=1, color='#4A86E8', width=0.008)

ax2.quiver(0, 0, Bx, By, angles='xy', scale_units='xy', scale=1, color='#34A853', width=0.005, alpha=0.3)
ax2.text(Bx + 0.1, By, 'B (asli)', color='#34A853', alpha=0.6, fontsize=9)

mBx, mBy = -Bx, -By
ax2.quiver(0, 0, mBx, mBy, angles='xy', scale_units='xy', scale=1, color='#8FCE00', width=0.008)

Rx2, Ry2 = Ax - Bx, Ay - By
ax2.quiver(0, 0, Rx2, Ry2, angles='xy', scale_units='xy', scale=1, color='#E69138', width=0.008)

ax2.plot([Ax, Rx2], [Ay, Ry2], linestyle='--', color='#8FCE00', alpha=0.8, linewidth=2)
ax2.plot([mBx, Rx2], [mBy, Ry2], linestyle='--', color='#4A86E8', alpha=0.6, linewidth=2)

ax2.text(Ax/2 + 0.1, Ay/2, 'A', color='#4A86E8', fontweight='bold', fontsize=12)
ax2.text(mBx/2 - 0.4, mBy/2 - 0.2, '-B', color='#8FCE00', fontweight='bold', fontsize=12)
ax2.text(Rx2/2 - 0.6, Ry2/2 + 0.1, 'A-B', color='#E69138', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()
```

- for `addition`, let say u have vector A(0.5, -1) & B(1, 1.5), u can as simple add x with x and y with y => (1.5,0.5). If u want to operate in other way u can do it by geometric approach, take the head of vector A and "collide" with nail vector B
- for `substraction` is simmilar, u can as simple substract like basic algebra. In geometry, subtraction can be interpreted as adding the opposite vector, or equivalently, reversing the direction and then performing the addition.

# Cosine Simillarity

```python
mag_A = 3.0
mag_B = 2.5
angle_A_deg = 15
theta_diffs = [0, 60, 90, 180]
titles = ["Identic (Cos = 1)", "Pointing in same direction (Cos = 0.5)", "Orthogonal (Cos = 0)", "Opposite (Cos = -1)"]

fig, axs = plt.subplots(2, 2, figsize=(14, 12))
axs = axs.flatten()

for i, theta in enumerate(theta_diffs):
    ax = axs[i]
    
    angle_B_deg = angle_A_deg + theta
    rad_A = np.radians(angle_A_deg)
    rad_B = np.radians(angle_B_deg)
    rad_theta = np.radians(theta)
    
    Ax = mag_A * np.cos(rad_A)
    Ay = mag_A * np.sin(rad_A)
    Bx = mag_B * np.cos(rad_B)
    By = mag_B * np.sin(rad_B)
    
    dot_prod = mag_A * mag_B * np.cos(rad_theta)
    cos_sim = np.cos(rad_theta)
    
    proj_mag = mag_B * np.cos(rad_theta)
    proj_x = proj_mag * np.cos(rad_A)
    proj_y = proj_mag * np.sin(rad_A)
    
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.3)
    
    ax.quiver(0, 0, Ax, Ay, angles='xy', scale_units='xy', scale=1, color='#4A86E8', width=0.008, zorder=3)
    
    if theta == 0:
        ax.quiver(0, 0, Bx, By, angles='xy', scale_units='xy', scale=1, color='#34A853', width=0.012, alpha=0.5, zorder=2)
    else:
        ax.quiver(0, 0, Bx, By, angles='xy', scale_units='xy', scale=1, color='#34A853', width=0.008, zorder=3)
    
    if theta not in [0, 180]:
        ax.plot([Bx, proj_x], [By, proj_y], linestyle='--', color='#E69138', alpha=0.8, linewidth=1.5)
    
    ax.plot(proj_x, proj_y, marker='o', color='#E69138', markersize=7, zorder=4)
    ax.text(proj_x, proj_y + 0.3, 'proj', color='#E69138', fontweight='bold', ha='center', fontsize=10)
    
    ax.text(Ax/2 + 0.2, Ay/2 - 0.2, 'A', color='#4A86E8', fontweight='bold', fontsize=12)
    if theta == 0:
        ax.text(Bx/2 - 0.3, By/2 + 0.2, 'B', color='#34A853', fontweight='bold', fontsize=12)
    elif theta == 180:
        ax.text(Bx/2, By/2 + 0.2, 'B', color='#34A853', fontweight='bold', fontsize=12)
    else:
        ax.text(Bx/2 - 0.2, By/2 + 0.2, 'B', color='#34A853', fontweight='bold', fontsize=12)
        
    ax.set_title(titles[i], pad=15, fontweight='bold', fontsize=13)
    
    info_text = (
        f"Angle = {theta}°\n"
        f"Dot Product = {dot_prod:.2f}\n"
        f"Cosine Sim = {cos_sim:.2f}"
    )
    props = dict(boxstyle='round,pad=0.6', facecolor='white', alpha=0.9, edgecolor='lightgray')
    ax.text(0.03, 0.97, info_text, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', bbox=props, color='#333333')

plt.tight_layout()
plt.show()
```

# Summary

Simply put. we can unstand `vector` as a point in space it can 2D,3D or even 768D like we often find in emebdding transformers.
1. Vector in $R^n$ <br>
    - A vector $v \in R^n$ is point/arrow in n-dimensional space:<br>
    - e.g. ridwan = [0.82, 0.91, -0.12, 0.45, 0.73, ...]  ← 768 dimensional vec
2. Vector operations
    - Addition: $u + v$ -> "word arithmetic": king − man + woman ≈ queen
    - Scalar multiplication: $\alpha v$ -> "zoom in/out" the vektor
    - Dot product: $u · v = \Sigma uᵢvᵢ = ‖u‖‖v‖cos(θ)$ -> measure how aligned two vector (crucial operation nowadays)
3. Magnitude/direction
    - $‖v‖ = √(v₁² + v₂² + ... + vₙ²)$
    - Unit vector: $v̂ = v / ‖v‖$ -> just direction, without magnitude. "Root" of cosine similarity.
4. Cosine Simillarity
    - $cosine(A, B) = \frac{(A · B)}{(‖A‖ · ‖B‖)}$
    - Range [-1, 1]:
        - 1 = identic
        - 0 = unrelated
        - -1 = opposite

# Example (code)

```python
cat = np.array([0.82, 0.91, -0.12, ])
lion = np.array([0.79, 0.88, -0.08, ])
car = np.array([-0.91, 0.15, 0.82, ])

# Dot product
print(np.dot(cat, lion))   
print(np.dot(cat, car))   

# Cosine similarity
def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

print(cosine_sim(cat, lion))  
print(cosine_sim(cat, car))  
```
